```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
def RK4(f,t0,T,y0,N):
    h = T/N
    tt = [t0+i*h for i in range(N+1)]
    Y =[y0]
    for i in range(1,N+1):
        k1 = f(tt[i-1],Y[i-1])
        k2 = f(tt[i-1]+h/2,Y[i-1]+k1*h/2)
        k3 = f(tt[i-1]+h/2,Y[i-1]+k2*h/2)
        k4 = f(tt[i-1]+h,Y[i-1]+k3*h)
        Y.append(Y[i-1]+h/6*(k1+2*(k2+k3)+k4))
    return tt,Y
```

```python
#Q3
a,b,c,p,k =0.8,0.5,0.09,0.004,0.005 
F = lambda t,X : np.array([-a*X[0]*X[2]+p*(sum(X))-k*X[0]  ,
                          a*X[0]*X[2]-(b+k)*X[1],
                          b*X[1]-(c+k)*X[2],
                           c*X[2]-k*X[3]])
```

```python
t0=0
N =10**3
T =80
y0 = np.array([0.6,0.1,0.3,0])
Tt,Y = RK4(F,t0,T,y0,N)
```

```python

S = [Y[i][0] for i in range(len(Y))]
E = [Y[i][1] for i in range(len(Y))]
I = [Y[i][2] for i in range(len(Y))]
R = [Y[i][3] for i in range(len(Y))]
N = [S[i]+E[i]+I[i]+R[i] for i in range(len(Y))]
```

```python
pl.plot(Tt,S)
pl.plot(Tt,E)
pl.plot(Tt,I)
pl.plot(Tt,R)
pl.plot(Tt,S)
pl.plot(Tt,N)
```
