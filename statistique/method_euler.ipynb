```python
def euler(f,a,b,y0,N):
    h = (b-a)/N
    T=[a+i*h for i in range(N+1)]
    Y=[y0]
    for i in range(1,N+1):
        Y.append(Y[i-1]+h*f(T[i-1], Y[i-1]))

    return (T,Y)
```

```python
#Q1
F = lambda t,u: u
a,b = 0,3
N = 100
u0 = 1
T,Y = euler(F,a,b,u0,N)
```

```python
def euler2(f,a,b,y0,v0,N):
    h = (b-a)/N
    T=[a+i*h for i in range(N+1)]
    Y=[y0]
    V=[v0]
    for i in range(1,N+1):
        Y.append(Y[i-1]+h*V[i-1])
        V.append(V[i-1]+h*f(T[i-1],Y[i-1],V[i-1]))
    return T,Y
```

```python
#Solution euler
F=lambda t,u,v:(2/t)-(v/t)+(u/t**2)
a,b=1,10
u0=1
v0=1
N=100
T,Y=euler2(F,a,b,u0,v0,N)
```

```python
#Solution theorique
s=lambda x:x/2+1/(2*x)+x*np.log(x)
S=[s(x) for x in T]
```

```python
#tracage des courbes
plt.plot(T,Y,'r')
plt.plot(T,S,'g')
```
