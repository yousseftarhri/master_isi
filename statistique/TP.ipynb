```python
#Q3
def euler( f,t0,y0,T,N):
    h=T/N
    tt=[t0+i*h for i in range(N+1)]
    Y=[y0]
    for i in range(1,N+1):
        Y.append(Y[i-1]+h*f(tt[i-1],Y[i-1]))
    return (tt,Y)
```

```python
#Q6
import numpy as np
F=lambda t,X:np.array([X[1],-4*(np.pi**2)*np.cos(2*np.pi*t)-X[0]-X[1]])
```

```python
#Q7
N=100
t0=0
T=3
u0=0
z0=0
Th,Yheun=heun(F,t0,np.array([u0,z0]),T,N)
Te,Yeuler=euler(F,t0,np.array([u0,z0]),T,N)

Yheun_=[Yheun[i][0] for i in range(len(Yheun))]
Yeuler_=[Yeuler[i][0] for i in range (len(Yeuler))]
```

```python
import matplotlib.pyplot as pt
pt.plot(Th,Yheun_,'r')
pt.plot(Te,Yeuler_,'g')
```
